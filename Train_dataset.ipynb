```python
import pandas as pd
import os
import ast
```

### NGOs

```python
path='datasets/extended_features.csv'
```

```python
df=pd.read_csv(path,dtype='str',index_col=0)
df['text_new']=df['text_new'].apply(ast.literal_eval)
df=df[~df.id.isin(['902203267001851904','1009290406108905473','1213468426062254082'])].copy()
df.index=df.index.astype(int)
df=df.dropna(subset='label').copy()
```

```python
df_out=pd.DataFrame(columns=df.columns)
df_hiddens = pd.DataFrame(columns=df.columns)
```

```python
pd.merge(df.groupby(["relatedNGO","label"]).count()[['conversation_id']],df.groupby(["relatedNGO","label"]).count()[['conversation_id']]/df.groupby(['relatedNGO']).count()[['conversation_id']],how='left',left_index=True,right_index=True)
```

```python
for ngo in ["redcrosscanada","philredcross","girlscouts","WorldVision","SalvationArmyUS","RedCrossAU","NepalRedCross","ICRC_Nairobi","Habitat_org","BritishRedCross"]:
    temp_df=df[df['relatedNGO']==ngo].copy()
    num_answers=temp_df[temp_df['label']=='1'].shape[0]
    df_out=pd.concat([df_out,temp_df[temp_df['label']=='1'].sample(frac=1)])
    # df_out=pd.concat([df_out,temp_df[temp_df['label']=='0'][:num_answers].sample(frac=1)])
    df_out=pd.concat([df_out,temp_df[temp_df['label']=='0'].sample(frac=1)])
```

```python
df_out.groupby('label').count()
```

```python
name='ngos'

df_out.sample(frac=0.8,random_state=0).to_csv(f"data/train_wa_{name}.csv")
df_out.sample(frac=0.15,random_state=0).to_csv(f"data/eval_wa_{name}.csv")
df_out.sample(frac=0.05,random_state=0).to_csv(f"data/test_wa_{name}.csv")
df_out[df_out["label"]=="1"].sample(frac=1,random_state=0).to_csv(f"data/oo_wa_{name}.csv")
df_out[df_out["label"]=="0"].sample(min(10000,df_out[df_out["label"]=="0"].shape[0]),random_state=0).to_csv(f"data/oz_wa_{name}.csv")
```

```python
df_unseen=df[~df['id'].isin(df_out['id'].values)].copy()
df_unseen.sample(10000,random_state=0).to_csv(f"data/la_10k_{name}.csv")
df_unseen[df_unseen['label']=='1'].to_csv(f"data/oo_la_{name}.csv")
df_unseen[df_unseen['label']=='0'].sample(10000,random_state=0).to_csv(f"data/oz_la_10k_{name}.csv")
```
